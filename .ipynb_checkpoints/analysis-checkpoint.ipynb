```python
import numpy as np
import matplotlib.pyplot as plt

import os, sys
```

```python
dat = np.loadtxt('_data/flux/outputs.dat')
```

```python
W = dat[:, -1]

dC = (dat[:, 5] - dat[:, 3])/(dat[:, 5] + dat[:, 3])
dP = (dat[:, 4] - dat[:, 2])/(dat[:, 4] + dat[:, 2])
```

```python
W1 = []
W2 = []

for i in range(len(dat)) :
    w = W[i]
    dc = dC[i]
    dp = dP[i]
    if w == 1 :
        W1 += [[dc, dp]]
    elif w == 2 :
        W2 += [[dc, dp]]
        
W1, W2 = np.array(W1), np.array(W2)
```

```python

chi= .01

def Lambda(chi) :
    return -(1-chi*np.sinh(1./chi))/(2.*chi*np.sinh(1./chi)) * np.cosh(0.5/chi)

def func_P2(P1, C1, C2, chi) :
    return P1 + (C2-C1)*Lambda(chi) / np.cosh(1./chi)

def func_P1(P2, C1, C2, chi) :
    return P2 - (C2-C1)*Lambda(chi) / np.cosh(1./chi)

P1 = 1.
C1 = 1.
C2 = np.linspace(1., 100., 101)

P2 = func_P2(P1, C1, C2, chi)
    
Delta_C2 = (C2-C1)/(C2+C1)
Delta_P2 = (P2-P1)/(P2+P1)

P2 = 1.
C2 = 1.
C1 = np.linspace(1., 100., 101)

P1 = func_P1(P2, C1, C2, chi)

Delta_C1 = (C2-C1)/(C2+C1)
Delta_P1 = (P2-P1)/(P2+P1)
```

```python
plt.scatter(W1[:, 0], W1[:, 1], color = 'b', label = '1 wins', marker='+', alpha=0.1)
plt.scatter(W2[:, 0], W2[:, 1], color = 'r', label = '2 wins', marker='+', alpha=0.1)
plt.plot(Delta_C2, Delta_P2, color='k', linewidth=4)
plt.plot(Delta_C1, Delta_P1, color='k', linewidth=4)

x = np.linspace(-1., 1., 1001)
alpha=0.3
plt.plot(x, alpha*x, linestyle='--', color='k')

plt.axis('equal')
plt.legend()
plt.grid()
plt.show()
```
