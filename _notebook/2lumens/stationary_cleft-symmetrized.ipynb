# Stationary solutions - Cleft
## The system
### Differential equations
We have the following differential equations for pressure and concentration in stationary state in the cleft :
$$\tag{1}\begin{equation}\chi_s^2 \frac{\partial^2 c}{\partial x^2} = c(x) - c^a\end{equation}$$
$$\tag{2}\begin{equation}\chi_v^2\frac{\partial^2 p}{\partial x^2} = p(x) - c(x)\end{equation}$$
where $\chi_v (t)= \xi_v / \ell(t)$, $\chi_s (t)= \xi_s / \ell(t)$, $c^a$ are (spatial) constants.
### Boundary conditions
We have four BC :
$$\begin{cases}
c(x=-1/2) &= \delta C_1 (= C_1 - 1)\\
c(x=1/2) &= \delta C_2 (= C_2 - 1) \\
p(x=-1/2) &= P_1 \\
p(x=1/2) &= P_2
\end{cases}$$
## Solutions
### Concentration
The solution for the concentration equation (1) is 
$$\tag{3}c(x) = c^a - (\delta C_1 - c^a) \frac{\sinh \left[\frac{x-\frac{1}{2}}{\chi_s} \right]}{\sinh(1/\chi_s)} + (\delta C_2 - c^a) \frac{\sinh \left[\frac{x+\frac{1}{2}}{\chi_s} \right]}{\sinh(1/\chi_s)}$$

### Pressure
The solution for the pressure equation (2) is
$$\tag{4}p(x) = -\frac{P_1 \sinh\left[\frac{x-\frac{1}{2}}{\chi_v}\right]}{\sinh (1/\chi_v)} + \frac{P_2 \sinh\left[\frac{x+\frac{1}{2}}{\chi_v}\right]}{ \sinh (1/\chi_v)} + \lambda(x)e^{x/\chi_v} + \mu(x)e^{-x/\chi_v}  + \frac{e^{-1/2\chi_v}}{\sinh 1/\chi_v} \left( \lambda(-\frac{1}{2}) \sinh \left[\frac{x-\frac{1}{2}}{\chi_v} \right] - \mu(\frac{1}{2}) \sinh \left[\frac{x+\frac{1}{2}}{\chi_v} \right] \right)$$
where
$$\lambda(x) = \int_{1/2}^x dy E^- (y)$$
$$\mu(x) = \int_{-1/2}^x dy E^+ (y)$$
$$E^\pm(x) = \pm \frac{e^{\pm x / \chi_v}}{2\chi_v} c \left[x ; \delta C_1, \delta C_2, \chi_s, c^a\right]$$

Analytical expressions of $\lambda(-1/2)$, $\mu(1/2)$, $\lambda(x)$ and $\mu(x)$ are quite cumbersome and given in Appendix

**N.B. :** $\lambda(1/2) = \mu(-1/2) = 0$

## Boundary Fluxes
## Solute fluxes $J_{1, 2}^s$

The solute fluxes are given by

$$J_1^s = - X_s^2 \partial_x c(x) |_{x=-1/2}$$
$$J_2^s = + X_s^2 \partial_x c(x) |_{x=1/2}$$
where 
$$X_s^2(t) = \frac{\xi_s^2}{L_0\ell(t)} = \chi_s \frac{\xi_s}{L_0}$$
So using the solution of $c$, we have

### Non-symmetrized and symmetrized
$$\tag{6.1}J_1^s = \chi_s \frac{\ell(t)}{L_0} \left[ (\delta C_1-c^a)\coth \left(\frac{1}{\chi_s}\right) - (\delta C_2-c^a)\frac{1}{\sinh\left( \frac{1}{\chi_s}\right)} \right]$$

$$\tag{6.2}J_2^s = \chi_s \frac{\ell(t)}{L_0} \left[ (\delta C_2-c^a)\coth \left(\frac{1}{\chi_s}\right) - (\delta C_1-c^a)\frac{1}{\sinh\left( \frac{1}{\chi_s}\right)} \right]$$


## Solvant fluxes $J_{1, 2}^v$

$$J_1^v = - X_v^2 \partial_x p(x) |_{x=-1/2}$$
$$J_2^v = + X_v^2 \partial_x p(x) |_{x=1/2}$$
where
$$X_v^2(t) = \frac{\xi_v^2}{L_0\ell(t)} = \frac{\ell(t)\chi_v^2}{L_0}$$

### Non-symmetrized
$$\tag{7.1}J_1^v = -\chi_v \frac{\ell(t)}{L_0} \left[ \coth \left( \frac{1}{\chi_v} \right)\left(\lambda(0) + \mu(0) - P_1 \right) + \frac{P_2}{\sinh(1/\chi_v)} +\lambda(0)-\mu(0)\right]$$
$$\tag{7.2}J_2^v = \chi_v \frac{\ell(t)}{L_0} \left[ \frac{1}{\sinh(1/\chi_v)} \left(\lambda(0) + \mu(0) - P_1 \right) + P_2\coth \left(\frac{1}{\chi_v}\right)\right] $$

### Symmetrized
$$\tag{8}J_1^v = \chi_v \frac{\ell(t)}{L_0} \left[ \coth \left( \frac{1}{\chi_v} \right)\left(P_1 - \lambda \left(-1/2\right) e^{-1/2\chi_v} \right) - \frac{1}{\sinh(1/\chi_v)} \left(P_2 - \mu \left(1/2\right) e^{-1/2\chi_v} \right) + \lambda \left(-1/2 \right) e^{-1/2\chi_v}\right]$$

$$\tag{9}J_2^v = \chi_v \frac{\ell(t)}{L_0} \left[ \coth \left( \frac{1}{\chi_v} \right)\left(P_2 - \mu \left(1/2\right) e^{-1/2\chi_v} \right) - \frac{1}{\sinh(1/\chi_v)} \left(P_1 - \lambda \left(-1/2\right) e^{-1/2\chi_v} \right) - \mu \left(1/2 \right) e^{-1/2\chi_v}\right] $$

# Local fluxes
## Solute flux

The local solute flux is found to be
$$J_s (x) \equiv - \frac{\xi_s}{L_0} \chi_s\frac{\partial c}{\partial x}$$
so that
$$J_s (x) = \frac{\xi_s}{L_0 \sinh\left[ \frac{1}{\chi_s} \right]}\left[(\delta C_1 - c^a) \cosh \left( \frac{x -\frac{1}{2}}{\chi_s} \right) - (\delta C_2 - c^a) \cosh \left( \frac{x +\frac{1}{2}}{\chi_s} \right)\right]$$

## Water flux

The local hydraulic is found to be
$$J_v (x) \equiv - \frac{\xi_v}{L_0}\chi_v\frac{\partial p}{\partial x}$$
so that
$$J_v(x) = \frac{\xi_v}{L_0} \left\lbrace \frac{1}{\sinh 1/\chi_v} \left[ (P_1-\lambda(\frac{1}{2})) \cosh\frac{x-\frac{1}{2}}{\chi_v} - (P_2 - \mu(\frac{-1}{2}) ) \cosh\frac{x+\frac{1}{2}}{\chi_v}\right]  - \lambda(x) e^{x/\chi_v} + \mu(x) e^{-x/\chi_v}\right\rbrace$$

# Dynamical equations for the lumens

The dynamical equations for the lumens $1, 2$ are given by dynamic of the length $L_j(t)$ of the lumen $j=1, 2$

$$\tag{10} \tau_v \frac{d L_j}{dt} = \mu_j\nu_j\left[ \left(\frac{\mu_j N_j}{L_j^2} - 1\right) -\frac{\epsilon_j}{L_j}\right] - \frac{\mu_j}{2L_j}J_j^v$$

but $\epsilon_j \ll 1$, and the dynamic of the number of ions $N_j(t)$

$$\tag{11} \tau_s \frac{d N_j}{dt} = 2\nu_j L_j \left[ 1 - \mu_j\frac{N_j}{L_j^2} + c^a \right] - J_j^s$$

Concentration $C_j$ and pressure $P_j$ in lumen $j$ are related to the equations via

$$ C_j(t) = \frac{N_j(t)}{A_j(t) c_0} = \mu_j \frac{N_j(t)}{L_j^2(t)}$$

$$ \delta C_j(t) = C_j - 1 = \mu_j \frac{N_j(t)}{L_j^2(t)} - 1$$

$$P_j(t) = \frac{\Gamma_j}{A_j^{1/2}(t)} = \frac{L_0}{L_j(t)}\epsilon_j$$
where $L = L_1(t) + L_2(t) + \ell(t) = L_1(0) + L_2(0) + \ell(0)$

```python
# Winning Diagram
def import_file(filename, nmax=2) :
    f = open(filename, 'r')
    L = f.readlines()
    
    dat = np.zeros((len(L)-1, len(L[0].split(';'))))
    for i in range(len(L[:nmax])) : 
        s = L[i+1].split(';')
        for n in range(len(s)) :
            s[n] = s[n].replace(',', '.')
            dat[i, n]  = float(s[n])            
    return dat
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

from ipywidgets import interactive
from scipy.integrate import quad

def c(x, chis, dC1, dC2, ca=0) :
    """
    c(x, chis, dC1, dC2, ca=0)
    
        Calculate the concentration difference between external medium and cleft
        at point x in the intercellular junction, given the constants.
    
        Parameters
        ---------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        chis : float
            Ratio of lengths : chis = xi_s / ell
            where xi_s is the osmolyte screening length, ell is the length of the intercellular junction.
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
    
        Returns
        -------
        c : float
            Value of the concentration at the given point x.
            NB : here, c is a concentration difference, ie : c(x) = c_cleft(x) - c0, so it can be negative !
    """
    c1 = (dC1 - ca) * np.sinh((x-0.5)/chis) / np.sinh(1./chis)
    c2 = (dC2 - ca) * np.sinh((x+0.5)/chis) / np.sinh(1./chis)
    return ca - c1 + c2

def p(x, chiv, P1, P2, chis, dC1, dC2, ca=0) :
    """
    p(x, chiv, P1, P2, chis, dC1, dC2, ca=0)
    
        Calculate the pressure difference between external medium and cleft
        at point x in the intercellular junction, given the constants.
    
        Parameters
        ---------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        chiv : float
            Ratio of lengths : chiv = xi_v / ell
            where xi_v is the hydraulic screening length, ell is the length of the intercellular junction.
        P1, P2 : float
            Pressure differences at the intercellular junction - lumen interfaces (1 and 2 respectively)
            due to Laplace's law. 
        chis : float
            Ratio of lengths : chis = xi_s / ell
            where xi_s is the osmolyte screening length, ell is the length of the intercellular junction.
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
       
        Returns
        -------
        p : float
            Value of the concentration at the given point x.
            NB : here, c is a concentration difference, ie : c(x) = c_cleft(x) - c0, so it can be negative !
        
        Requires
        --------
        lam_v2, ,u_v2 : functions
           Functions that analytically solve the ode for p (see associated notes).
    """
    p1 = P1 * np.sinh((x-0.5)/chiv) / np.sinh(1./chiv)
    p2 = P2 * np.sinh((x+0.5)/chiv) / np.sinh(1./chiv)
    
    lambda_x = lam_v2(x, chiv, chis, dC1, dC2, ca) * np.exp(x / chiv)
    mu_x = mu_v2(x, chiv, chis, dC1, dC2, ca) * np.exp(-x / chiv)

    lambda_a = np.exp(-0.5/chiv)*lam_v2(-0.5, chiv, chis, dC1, dC2, ca) * np.sinh((x-0.5)/chiv) / np.sinh(1./chiv)

    mu_b = np.exp(-0.5/chiv)*mu_v2(0.5, chiv, chis, dC1, dC2, ca) * np.sinh((x+0.5)/chiv) / np.sinh(1./chiv)

    return - p1 + p2 + lambda_x + mu_x + lambda_a - mu_b

# ===== Version 2 : solved with Wolfram, does work

def I_1_minus_v2(x, Xv, Xs) :
    """
    I_1_minus_v2(x, Xv, Xs)
        
        Function used to solve analytically lambda (the first term). Given by Wolframalpha.
        NB : _1_ refers to the first integral term in the integral, associated to dC1
            _minus_ refers to the minus in the exponential term in the integral of lambda.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        NB : Xv = Xs is a special case !
    
        Returns
        -------
        I : float
            Value of the integral at point x.
    """
    c1, s1 = np.cosh(1./Xs), np.sinh(1./Xs)
    cmx, smx = np.cosh((x-0.5)/Xs), np.sinh((x-0.5)/Xs)
    
    if Xv == Xs :
        return 0.5*( (x-0.5)*np.exp(-0.5/Xv) + 0.5*Xv*(np.exp(-2.*x/Xv) - np.exp(-1./Xv))*np.exp(0.5/Xv) )
    
    return Xv*Xs*(np.exp(-x/Xv)*(Xv*cmx + Xs*smx) - Xv*np.exp(-0.5/Xv))/ ( (Xv-Xs)*(Xv+Xs) )

def I_2_minus_v2(x, Xv, Xs) :
    """
    I_2_minus_v2(x, Xv, Xs)
        
        Function used to solve analytically lambda (the first term). Given by Wolframalpha.
        NB : _2_ refers to the second integral term in the integral, associated to dC2
            _minus_ refers to the minus in the exponential term in the integral of lambda.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        NB : Xv = Xs is a special case !
    
        Returns
        -------
        I : float
            Value of the integral at point x.
    """
    c1, s1 = np.cosh(1./Xs), np.sinh(1./Xs)
    cpx, spx = np.cosh((x+0.5)/Xs), np.sinh((x+0.5)/Xs)
    
    if Xv == Xs :
        return 0.5*( (x-0.5)*np.exp(0.5/Xv) + 0.5*Xv*(np.exp(-2.*x/Xv) - np.exp(-1./Xv))*np.exp(-0.5/Xv) )
    
    return Xv*Xs*(np.exp(-x/Xv)*(Xv*cpx + Xs*spx) - np.exp(-0.5/Xv)*(Xv*c1 + Xs*s1))/ ( (Xv-Xs)*(Xv+Xs) )

def I_1_plus_v2(x, Xv, Xs) :
    """
    I_1_plus_v2(x, Xv, Xs)
        
        Function used to solve analytically lambda (the first term). Given by Wolframalpha.
        NB : _1_ refers to the first integral term in the integral, associated to dC1
            _plus_ refers to the plus in the exponential term in the integral of mu.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        NB : Xv = Xs is a special case !
    
        Returns
        -------
        I : float
            Value of the integral at point x.
    """
    c1, s1 = np.cosh(1./Xs), np.sinh(1./Xs)
    cmx, smx = np.cosh((x-0.5)/Xs), np.sinh((x-0.5)/Xs)
    
    if Xv == Xs :
        return 0.5*( 0.5*Xv*(np.exp(2.*x/Xv) - np.exp(-1./Xv))*np.exp(-0.5/Xv) - (x+0.5)*np.exp(0.5/Xv))
    
    return Xv*Xs*(np.exp(x/Xv)*(Xv*cmx - Xs*smx) - np.exp(-0.5/Xv)*(Xv*c1 + Xs*s1))/ ( (Xv-Xs)*(Xv+Xs) )

def I_2_plus_v2(x, Xv, Xs) :
    """
    I_2_plus_v2(x, Xv, Xs)
        
        Function used to solve analytically lambda (the first term). Given by Wolframalpha.
        NB : _2_ refers to the second integral term in the integral, associated to dC2
            _plus_ refers to the plus in the exponential term in the integral of mu.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        NB : Xv = Xs is a special case !
    
        Returns
        -------
        I : float
            Value of the integral at point x.
    """
    c1, s1 = np.cosh(1./Xs), np.sinh(1./Xs)
    cpx, spx = np.cosh((x+0.5)/Xs), np.sinh((x+0.5)/Xs)
    
    if Xv == Xs :
        return 0.5*( 0.5*Xv*(np.exp(2.*x/Xv) - np.exp(-1./Xv))*np.exp(0.5/Xv) - (x+0.5)*np.exp(-0.5/Xv))
    
    return Xv*Xs*(np.exp(x/Xv)*(Xv*cpx - Xs*spx) - np.exp(-0.5/Xv)*Xv)/ ( (Xv-Xs)*(Xv+Xs) )

def lam_v2(x, Xv, Xs, dC1, dC2, ca=0) :
    """
    lam_v2(x, Xv, Xs, dC1, dC2, ca=0)
    
        Calculate the function lambda(x) in the pressure analytical solution.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
    
        Returns
        -------
        lam : float
            The value of lambda(x) at point x.
        
        Requires
        --------
        I_1_minus_v2, I_2_minus_v2 : functions
    
    """
    la = 0.5*ca * (np.exp(-x/Xv) - np.exp(-0.5/Xv))
    l1 = (dC1-ca) * I_1_minus_v2(x, Xv, Xs) / (2.*np.sinh(1./Xs))
    l2 = (dC2-ca) * I_2_minus_v2(x, Xv, Xs) / (2.*np.sinh(1./Xs))
    return la + l1 - l2

def mu_v2(x, Xv, Xs, dC1, dC2, ca=0) :
    """
    mu_v2(x, Xv, Xs, dC1, dC2, ca=0)
    
        Calculate the function mu(x) in the pressure analytical solution.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        Xv : float
            Ratio of hydraulic / cleft lengths
        Xs : float
            Ratio of osmotic / cleft lengths
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
    
        Returns
        -------
        mu : float
            The value of mu(x) at point x.
        
        Requires
        --------
        I_1_plus_v2, I_2_plus_v2 : functions
    
    """
    ma = 0.5*ca * (np.exp(x/Xv) - np.exp(-0.5/Xv))
    m1 = (dC1-ca) * I_1_plus_v2(x, Xv, Xs) / (2.*np.sinh(1./Xs))
    m2 = (dC2-ca) * I_2_plus_v2(x, Xv, Xs) / (2.*np.sinh(1./Xs))
    return ma - m1 + m2

# ==== System of functions

def func_Lj(Lj, Nj, tau_v, mu_j, nu_j, eps_j, Jjv) :        
    """
    func_Lj(L, N, tau_v, mu_j, nu_j, eps_j, Jjv)
        
        Function that calculates the length variation of lumen j.
    
        Parameters
        ----------
        Lj : float
            Length of the lumen j
        Nj : float
            Number of ions in lumen j
        tau_v : float
            Hydraulic caracteristic time
        mu_j : float
            Geometrical factor
        nu_j : float
            Geometrical factor
        eps_j : float
            Osmotic vs hydraulic pressure term for lumen j
        Jjv : float
            Outgoing hydraulic flux for lumen j. If positive, water goes outside the lumen.
        
        Returns
        -------
        dLjdt : float
            Length variation.
    """
    return (mu_j*nu_j*(mu_j * Nj / (Lj*Lj) - 1. - eps_j / Lj) - mu_j * Jjv / (2*Lj))/tau_v

def func_Nj(Lj, Nj, tau_s, mu_j, nu_j, ca, Jjs) :
    """
    func_Nj(Lj, Nj, tau_s, mu_j, nu_j, ca, Jjs)
        
        Function that calculates the variation fo the number of ions of lumen j.
    
        Parameters
        ----------
        Lj : float
            Length of the lumen j
        Nj : float
            Number of ions in lumen j
        tau_s : float
            Osmotic caracteristic time
        mu_j : float
            Geometrical factor
        nu_j : float
            Geometrical factor
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        Jsv : float
            Outgoing osmotic flux for lumen j. If positive, water goes outside the lumen.
        
        Returns
        -------
        dNjdt : float
            Nb of ions' variation.
    """
    return (2*nu_j*Lj*(1 + ca - mu_j*Nj/(Lj*Lj)) -  Jjs)/tau_s
    
# ==== Fluxes
## ==== Osmotic fluxes
def func_J1s(chis, ellt, L, dC1, dC2, ca) :
    """
    func_J1s(chis, ellt, L, dC1, dC2, ca)
    
        Outgoing osmotic flux of lumen 1.
    
        Parameters
        ----------
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        ellt : float
            Current length of the cleft, at time t. If this length is negative, there is a problem.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        
        Returns
        -------
        J1s : float
            Value of the outgoing flux of lumen 1. If positive, the flux is from lumen 1 to the cleft.
    """
    if ellt <= 0. :
        return 0.
    return chis*ellt/L * ((dC1-ca)*np.cosh(1./chis) - (dC2-ca))/np.sinh(1./chis)

def func_J2s(chis, ellt, L, dC1, dC2, ca) :
    """
    func_J2s(chis, ellt, L, dC1, dC2, ca)
    
        Outgoing osmotic flux of lumen 2.
    
        Parameters
        ----------
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        ellt : float
            Current length of the cleft, at time t. If this length is negative, there is a problem.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        
        Returns
        -------
        J2s : float
            Value of the outgoing flux of lumen 2. If positive, the flux is from lumen 2 to the cleft.
    """
    if L <= 0. :
        return 0.
    return chis*ellt/L * ((dC2-ca)*np.cosh(1./chis) - (dC1-ca)) / np.sinh(1./chis)

## ==== Hydraulic fluxes
def func_J1v(chiv, ellt, L, P1, P2, chis, dC1, dC2, ca) :
    """
    func_J1v(chiv, ellt, L, P1, P2, chis, dC1, dC2, ca)
    
        Outgoing hydraulic flux of lumen 1.
    
        Parameters
        ----------
        chiv : float
            Ratio of lengths : chiv = xi_v / ell
            where xi_v is the hydraulic screening length, ell is the length of the intercellular junction.
        ellt : float
            Current length of the cleft, at time t. If this length is negative, there is a problem.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        P1, P2 : floats
            Pressure differences at the intercellular junction - lumen interfaces (1 and 2 respectively)
            due to Laplace's law.
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        dC1, dC2 : floats
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        
        Returns
        -------
        J1v : float
            Value of the outgoing flux of lumen 1. If positive, the flux is from lumen 1 to the cleft.
    
        Requires
        --------
        lam_v2, mu_v2 : functions
            Functions that analytically solve the ode for p (see associated notes).
    """
    la = lam_v2(-0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    mb = mu_v2(0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    if L <= 0. :
        return 0.
    return chiv*ellt/L * ((P1-la)*np.cosh(1./chiv)/np.sinh(1./chiv) - (P2-mb)/np.sinh(1./chiv) - la)

def func_J2v(chiv, ellt, L, P1, P2, chis, dC1, dC2, ca) :
    """
    func_J2v(chiv, ellt, L, P1, P2, chis, dC1, dC2, ca)
    
        Outgoing hydraulic flux of lumen 2.
    
        Parameters
        ----------
        chiv : float
            Ratio of lengths : chiv = xi_v / ell
            where xi_v is the hydraulic screening length, ell is the length of the intercellular junction.
        ellt : float
            Current length of the cleft, at time t. If this length is negative, there is a problem.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        P1, P2 : floats
            Pressure differences at the intercellular junction - lumen interfaces (1 and 2 respectively)
            due to Laplace's law.
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        dC1, dC2 : floats
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        
        Returns
        -------
        J2v : float
            Value of the outgoing flux of lumen 2. If positive, the flux is from lumen 2 to the cleft.
    
        Requires
        --------
        lam_v2, mu_v2 : functions
            Functions that analytically solve the ode for p (see associated notes).
    """
    la = lam_v2(-0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    mb = mu_v2(0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    if L <= 0. :
        return 0.
    return chiv*ellt/L * ((P2-mb)*np.cosh(1./chiv)/np.sinh(1./chiv) - (P1-la)/np.sinh(1./chiv) - mb)

## ==== Local fluxes
def func_Js(x, dC1, dC2, chis, ca, xi_s, L) :
    """
    func_Js(x, dC1, dC2, chis, ca, xi_s, L)
    
        Osmotic flux in the intercellular junction.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        xi_s : float
            Osmotic screening length.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        
        Returns
        -------
        Js : float
            Value of the osmotic flux in x. If positive, the flux goes in the positive x-axis direction.
    """
    c1 = (dC1-ca) * np.cosh((x-0.5)/chis) / np.sinh(1/chis)
    c2 = (dC2-ca) * np.cosh((x+0.5)/chis) / np.sinh(1/chis)
    return (xi_s / L) * (c1-c2)

def func_Jv(x, P1, P2, chiv, dC1, dC2, chis, ca, xi_v, L) :
    """
    func_Jv(x, P1, P2, chiv, dC1, dC2, chis, ca, xi_v, L)
    
        Hydraulic flux in the intercellular junction.
    
        Parameters
        ----------
        x : float
            Position along the x-axis. Must be in the interval [-0.5, 0.5].
        P1, P2 : floats
            Pressure differences at the intercellular junction - lumen interfaces (1 and 2 respectively)
            due to Laplace's law.
        chiv : float
            Ratio of lengths : chiv = xi_v / ell
            where xi_v is the hydraulic screening length, ell is the length of the intercellular junction.
        dC1, dC2 : float
            Concentration differences at the intercellular junction - lumen interfaces (1 and 2 respectively).
            Defined as dCi = Ci - 1
            where Ci is the normalized concentration in lumen i, 1 is the external normalized concentration c0.
        chis : float
            Ratio of lengths : chis = xi_s / ellt
            where xi_s is the osmolyte screening length, ellt is the length of the intercellular junction.
        ca : float
            Active pumping term. Positive to pump inward, negative to pump outside the intercellular junction.
        xi_v : float
            Hydraulic screening length.
        L : float
            Total length of the system, such that L = L1 + L2 + ellt
        
        Returns
        -------
        Jv : float
            Value of the hydraulic flux in x. If positive, the flux goes in the positive x-axis direction.
    
        Requires
        --------
        lam_v2, mu_v2 : functions
            Functions that analytically solve the ode for p (see associated notes).
    """
    la = lam_v2(-0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    mb = mu_v2(0.5, chiv, chis, dC1, dC2, ca)*np.exp(-0.5/chiv)
    
    lam_x = lam_v2(x, chiv, chis, dC1, dC2, ca) * np.exp(x / chiv)
    mu_x = mu_v2(x, chiv, chis, dC1, dC2, ca) * np.exp(-x / chiv)
    
    p1 = (P1-la)*np.cosh((x-0.5)/chiv) / np.sinh(1./chiv)
    p2 = (P2-mb)*np.cosh((x+0.5)/chiv) / np.sinh(1./chiv)
    
    return (xi_v / L)*( p1 - p2 - lam_x + mu_x)
    
def func_C_j(Nj, Lj, mu_j) :
    """
    func_C_j(Nj, Lj, mu_j)
    
        Calculate the concentration of lumen j.
    
        Parameters
        ----------
        Nj : float
            Number of ions in lumen j
        Lj : float
            Length of the lumen j
        mu_j : float
            Geometrical factor
    
        Returns
        -------
        Cj : float
            Value of the concentration in lumen j.
    """
    return mu_j * Nj / (Lj*Lj)

def func_P_j(Lj, eps_j, Ltot) :
    """
    func_P_j(Lj, eps_j, Ltot)
    
        Calculate the pressure of lumen j.
    
        Parameters
        ----------
        Lj : float
            Length of the lumen j
        eps_j : float
            Osmotic vs hydraulic pressure term for lumen j
        Ltot : float
            Total length of the system.
    
        Returns
        -------
        Pj : float
            Value of the pressure in lumen j.
    """
    return Ltot*eps_j / Lj

def area(length, theta) :
    """
    area(length, theta)
    
        Gives the area as a function of the contact angle and the length of the lumen, assuming axisymmetry.
    
        Parameters
        ----------
        length : float
            Length of the lumen
        theta : float
            Contact angle.
    
        Returns
        -------
        A : float
            Area of the lumen.
    """
    return length**2 * (2*theta - np.sin(2*theta)) / np.sin(theta)**2

def length(area, theta) :
    """
    length(area, theta)
    
        Gives the length as a function of the contact angle and the area of the lumen, assuming axisymmetry.
    
        Parameters
        ----------
        area : float
            area of the lumen
        theta : float
            Contact angle.
    
        Returns
        -------
        L : float
            Length of the lumen.
    """
    mu = np.sin(theta)**2 / (2*theta - np.sin(2*theta))
    return np.sqrt(area*mu)
    
def syst(Y, t, xi_s, xi_v, tau_s, tau_v, ca, ell0, e0, mu_1, mu_2, nu_1, nu_2, eps_1, eps_2, Ltot, sim_type, status) :
    L1, N1, L2, N2 = Y
    lt = Ltot - L1 - L2
    Xs, Xv      = np.sqrt(xi_s**2 / (lt*Ltot)), np.sqrt(xi_v**2 / (lt*Ltot))
    chis, chiv  = np.sqrt(xi_s**2 / (lt*lt)), np.sqrt(xi_v**2 / (lt*lt))
    
    
    #if lt < 1e-6 :
    #    print 'Merge !'
    #print Ltot, L1, L2, lt
    if sim_type == 'normal' :
        a1, a2, a3, a4, a5, a6, a7, a8 = 1., 1., 1., 1., 1., 1., 1., 1.
    elif sim_type == 'no_flux' :
        a1, a2, a3, a4, a5, a6, a7, a8 = 1., 1., 1., 1., 0., 0., 0., 0.
    elif sim_type == 'cst_area' :
        a1, a2, a3, a4, a5, a6, a7, a8 = 0., 1., 0., 1., 1., 0., 1., 0.
    elif sim_type == 'cst_ion' :
        a1, a2, a3, a4, a5, a6, a7, a8 = 1., 0., 1., 0., 0., 1., 0., 1.
    
    
    if L1 >= e0*e0 and L2 >= e0*e0 :#and N1 >= 1e-6 and N2 >= 1e-6 and lt >= 1e-6:

        C1, C2      = func_C_j(N1, L1, mu_1), func_C_j(N2, L2, mu_2)
        dC1, dC2    = C1 - 1., C2 - 1.
        P1, P2      = func_P_j(L1, eps_1, Ltot), func_P_j(L2, eps_2, Ltot)
        
        J1s = func_J1s(chis, lt, Ltot, dC1, dC2, ca)
        J2s = func_J2s(chis, lt, Ltot, dC1, dC2, ca)
        J1v = func_J1v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
        J2v = func_J2v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
        
        ## Calculations
        
        Y1 = a1*func_Lj(L1, N1, tau_v, mu_1, nu_1, eps_1, J1v)                  # L1
        Y2 = a2*func_Nj(L1, N1, tau_s, mu_1, nu_1, ca, J1s)                     # N1
    
        Y3 = a3*func_Lj(L2, N2, tau_v, mu_2, nu_2, eps_2, J2v)                  # L2
        Y4 = a4*func_Nj(L2, N2, tau_s, mu_2, nu_2, ca, J2s)                     # N2

        dYdt = [Y1, Y2, Y3, Y4]
        
    else :
        if L1 < e0*e0 :
            C1, C2      = 1., func_C_j(N2, L2, mu_2)
            dC1, dC2    = C1 - 1., C2 - 1.
            P1, P2      = 1., func_P_j(L2, eps_2, Ltot)
            
            J1s = func_J1s(chis, lt, Ltot, dC1, dC2, ca)
            J2s = func_J2s(chis, lt, Ltot, dC1, dC2, ca)
            J1v = func_J1v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
            J2v = func_J2v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
            
            Y1 = 0.#a1*func_Lj(L1, N1, tau_v, mu_1, nu_1, eps_1, J1v)                  # L1
            Y2 = 0.#a2*func_Nj(L1, N1, tau_s, mu_1, nu_1, ca, J1s)                     # N1
    
            Y3 = a3*func_Lj(L2, N2, tau_v, mu_2, nu_2, eps_2, J2v)                  # L2
            Y4 = a4*func_Nj(L2, N2, tau_s, mu_2, nu_2, ca, J2s)                     # N2
            
            dYdt = [Y1, Y2, Y3, Y4]
            
            if L2 >= 1.-e0 :
                #print 'max reached'
                dYdt = [0., 0., 0., 0.]
                
        elif L2 < e0*e0 :
            C1, C2      = func_C_j(N1, L1, mu_2), 1.
            dC1, dC2    = C1 - 1., C2 - 1.
            P1, P2      = func_P_j(L1, eps_1, Ltot), 1.
            
            J1s = func_J1s(chis, lt, Ltot, dC1, dC2, ca)
            J2s = func_J2s(chis, lt, Ltot, dC1, dC2, ca)
            J1v = func_J1v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
            J2v = func_J2v(chiv, lt, Ltot, P1, P2, chis, dC1, dC2, ca)
            
            Y1 = a1*func_Lj(L1, N1, tau_v, mu_1, nu_1, eps_1, J1v)                  # L1
            Y2 = a2*func_Nj(L1, N1, tau_s, mu_1, nu_1, ca, J1s)                     # N1
    
            Y3 = 0.#a3*func_Lj(L2, N2, tau_v, mu_2, nu_2, eps_2, J2v)                  # L2
            Y4 = 0.#a4*func_Nj(L2, N2, tau_s, mu_2, nu_2, ca, J2s)                     # N2
            
            dYdt = [Y1, Y2, Y3, Y4]
            
            if L1 >= 1.-e0 :
                #print 'max reached'
                dYdt = [0., 0., 0., 0.]
        else :
            Y1, Y2, Y3, Y4 = 0., 0., 0., 0.
            dYdt = [Y1, Y2, Y3, Y4]
            
        status = ''
    return dYdt

       
def stop_sim(lt, e0) :
    b = True
    if lt < e0 :
        b = False
    return b
    
# ==== Other functions

def give_muj(thetaj) :
    return np.sin(thetaj)**2 / (2*thetaj - np.sin(2*thetaj))
    
def give_nuj(thetaj) :
    return thetaj/np.sin(thetaj)
    
def give_epsj(L0, pi0, thetaj, gammaj) :
    return gammaj * np.sin(thetaj) / (L0*pi0)
    
    
#
```
