```python
def read_eventfile(eventfile) :
    s = open(eventfile, 'r').readlines()
    
    if 'merge' in s[0] :
        return ('M', int(s[0].split()[-1]))
    elif 'disappears' in s[0] :
        return ('D', int(s[-1].split()[-1]))
    else : 
        return None
    
def read_configfile(config_folder, param) :
    #sl = open(os.path.join(config_folder, 'lumens.dat'), 'r').readlines()
    #sb = open(os.path.join(config_folder, 'bridges.dat'), 'r').readlines()
    lumens = np.loadtxt(os.path.join(config_folder, 'lumens.dat'))
    bridges = np.loadtxt(os.path.join(config_folder, 'bridges.dat'))
    print(lumens)
    print(bridges)
    if param = 'ca' :
        return lumens[1:-1, -1], bridges[1:-1, -1]
    elif param = 'length' :
        return lumens[1:-1, 2], bridges[1:-1, 3]
    elif param = 'nb_ions' :
        return lumens[1:-1, 3]
    return None
```

```python
eventfile = os.path.join(main_dir, 'chi'+'1', 'run0001', 'events.log')
config_folder = os.path.join(main_dir, 'chi'+'1', 'run0001', 'config')
```
