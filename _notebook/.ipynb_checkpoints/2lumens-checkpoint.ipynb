# Fluxes

```python
def plot_fluxes(subdir, ca='5.10^{-2}', chiv='1.10^{-1}') :
    conf, ch = chain.load_config(os.path.join(main_dir, subdir, 'config.conf'))
    #print(ch)
    x = np.linspace(0., ch.total_length, 1001)
    #tools.plot_profile(x, ch)

    ell = tools.load_brfile(os.path.join(main_dir, subdir, 'sim_bridges.dat'))

    x_log = 0

    if ch.lumen_type == 'hydroosmotic' :
        L, nions, pos = tools.load_file(os.path.join(main_dir, subdir, 'sim_all.dat'), hydroosmotic=True)
        #tools.plot_evolution(L, nions, ell, x_logscale=x_log)
    else :
        L, pos = tools.load_file(os.path.join(main_dir, subdir, 'sim_all.dat'), hydroosmotic=False)
        #tools.plot_evolution_hydraulic(L, ell, x_logscale=x_log)
        
    L_f = np.loadtxt(os.path.join(main_dir, subdir, 'fluxes_L.dat'))
    N_f = np.loadtxt(os.path.join(main_dir, subdir, 'fluxes_N.dat'))

    flux_lengths = np.zeros((len(L_f)//2, 5))
    flux_lengths[:, 0] = L_f[::2, 0]
    flux_lengths[:, 1] = L_f[::2, 2]
    flux_lengths[:, 2] = L_f[::2, 3]
    flux_lengths[:, 3] = L_f[1::2, 2]
    flux_lengths[:, 4] = L_f[1::2, 3]

    fig, ax = plt.subplots(1, 2, figsize=(9, 4))

    ax[0].plot(flux_lengths[:, 0], flux_lengths[:, 1], label = '$J_{lat}^1$', color = 'r', linestyle='-')
    ax[0].plot(flux_lengths[:, 0], -flux_lengths[:, 2], label = '$J_{exch}^1$', color = 'r', linestyle='--')

    ax[0].plot(flux_lengths[:, 0], flux_lengths[:, 3], label = '$J_{lat}^2$', color = 'b', linestyle='-')
    ax[0].plot(flux_lengths[:, 0], -flux_lengths[:, 4], label = '$J_{exch}^2$', color = 'b', linestyle='--')

    ax[0].legend(fontsize=12)
    ax[0].grid()
    ax[0].set_xlabel('Time', fontsize=15)
    ax[0].set_ylabel('Fluxes Hydro', fontsize=15)

    ax[1].plot(L[:, 0], L[:, 1], label = r'$L_1$', color = 'r')
    ax[1].plot(L[:, 0], L[:, 2], label = r'$L_2$', color = 'b')
    ax[1].legend(fontsize=15)
    ax[1].grid()
    ax[1].set_xlabel('Time', fontsize=15)
    ax[1].set_ylabel('Lengths', fontsize=15)

    plt.suptitle(r'$c^a_1 = '+str(ca)+'$, $\chi_v = '+str(chiv)+'$', fontsize=15)
    plt.show()
    
    flux_nions = np.zeros((len(N_f)//2, 5))
    flux_nions[:, 0] = N_f[::2, 0]
    flux_nions[:, 1] = N_f[::2, 2]
    flux_nions[:, 2] = N_f[::2, 3]
    flux_nions[:, 3] = N_f[1::2, 2]
    flux_nions[:, 4] = N_f[1::2, 3]

    fig, ax = plt.subplots(1, 2, figsize=(9, 4))

    ax[0].plot(flux_nions[:, 0], flux_nions[:, 1], label = '$J_{lat}^1$', color = 'r', linestyle='-')
    ax[0].plot(flux_nions[:, 0], -flux_nions[:, 2], label = '$J_{exch}^1$', color = 'r', linestyle='--')

    ax[0].plot(flux_nions[:, 0], flux_nions[:, 3], label = '$J_{lat}^2$', color = 'b', linestyle='-')
    ax[0].plot(flux_nions[:, 0], -flux_nions[:, 4], label = '$J_{exch}^2$', color = 'b', linestyle='--')

    ax[0].legend(fontsize=12)
    ax[0].grid()
    ax[0].set_xlabel('Time', fontsize=15)
    ax[0].set_ylabel('Fluxes Osmo', fontsize=15)

    ax[1].plot(nions[:, 0], nions[:, 1], label = r'$N_1$', color = 'r')
    ax[1].plot(nions[:, 0], nions[:, 2], label = r'$N_2$', color = 'b')
    ax[1].legend(fontsize=15)
    ax[1].grid()
    ax[1].set_xlabel('Time', fontsize=15)
    ax[1].set_ylabel(r'$N_{ions}$', fontsize=15)

    plt.suptitle(r'$c^a_1 = '+str(ca)+'$, $\chi_v = '+str(chiv)+'$', fontsize=15)
    plt.show()
    
```

```python
plot_fluxes('fluxes/chiv1e-01', ca=0.05, chiv = 0.1)
```
