```python
def calc_mu(theta) :
    return np.sin(theta)**2 / (2*theta - np.sin(2*theta))

def calc_chi(theta, gamma, kappa, ell0, L0) :
    mu = calc_mu(theta)
    return 0.5*mu*np.sin(theta)*gamma*kappa / (ell0*L0**3)
    #return gamma*kappa / (ell0*L0**3)

def calc_chi(theta, eps, kappa, ell0, L0) :
    mu = calc_mu(theta)
    return 0.5*mu*np.sin(theta)*eps*kappa / (ell0*L0**3)
    #return mu*np.sin(theta)*eps / (L0*ell0**3)

```

```python
def lin(x, a, b) :
    return a*x+b

def fit_lin(t, N) :
    x, y = np.log(t), np.log(N)
    popt, pcov = curve_fit(lin, x, y)
    alpha, kappa = popt[0], np.exp(popt[1])
    alpha_std, kappa_std = pcov[0, 0], np.exp(pcov[1, 1])
    return kappa, alpha#, kappa_std, alpha_std
    
```

# Number of lumens

For RK45 integrations, we choose an evolutive time step, set as
$$\Delta t (t) = \frac{\alpha}{N(t)^{\beta}} $$
We compare the choice of $\beta$, which is arbitrary, to RKF45 integration. The best choice seems to be $\beta \sim 3$, as a trade-off between precision (difference to the RKF45 result) and rapidity (number of steps needed to reach the final step).
Quantitative results still need to be provided.

# Coarsening

```python
def plot_coarsening(tau, ell, L_avg, nsim, path, directory='run', color='k', theta=np.pi/3., gamma=1., kappa=1., eps=1e-3, chi_bool=True, fit=False, indices_fit = [0, -1], plot_powerlaw=True, t_powlaw=[-1, 1]) :
    has_legend = False
    a_list, k_list = [], []
    Nt_list = {}
    T_val = 1./calc_chi(theta, eps, kappa, ell0=ell, L0=L_avg)
    
    if chi_bool :
        #L0_1 = ell_1 + 2*L_avg
        #chi_val = calc_chi(theta, gamma, kappa, ell0=ell, L0=L_avg)
        
        val = T_val/(tau)
        plt.axvline(x=val, color = color, label = r'$\chi_{'+str(ell)+'}^{-1}=$'+"{:1.3e}".format(val))
   
    for i in range(nsim) :
        if 1 :
        #try :
            sub_path = os.path.join(path, directory+str(i).zfill(4))
            Nt = np.loadtxt(os.path.join(sub_path, 'sim_nlum.dat'))
            Nt_list[i] = Nt
            
            if fit :
                start, end = indices_fit[0], indices_fit[1]
                k, a = fit_lin(Nt[start:end, 0], Nt[start:end, 1])
                a_list += [a]
                k_list += [k]

            if not has_legend :
                plt.plot(Nt[:, 0]/T_val, Nt[:, 1], alpha=1, color=color, label = r'$\bar{\ell}(0) \sim '+str(ell)+'$', marker='o', markersize=1, linewidth=0)
                has_legend=True
            else :
                plt.plot(Nt[:, 0]/T_val, Nt[:, 1], alpha=0.2, color=color, marker='o', markersize=1, linewidth=0)
        
        
        #except :
        #    print(i)
        #    pass
            
    if fit :
        k_avg = np.average(k_list)
        a_avg = np.average(a_list)
        #t_pwlw = np.logspace(t_powlaw[0], t_powlaw[1], 101)
        t_pwlw = Nt[start:end, 0]
        if plot_powerlaw :
            plt.plot(t_pwlw, k_avg*t_pwlw**a_avg, color = 'yellow')
    
        return k_avg, a_avg, Nt_list
    else : return 0., 0., Nt_list
```

```python
def gen_log_times(tmin, tmax, npts) :
    return np.logspace(np.log10(tmin), np.log10(tmax), npts)

def find_argmins(Nt, log_times) :
    index_array = np.zeros(len(log_times), dtype=int)
    for k in range(len(log_times)) :
        index_array[k] = np.argmin(np.abs(log_times[k]-Nt[:, 0]))
    return index_array

def gen_index_array(Nt, npts) :
    size = len(Nt)
    
    tmin = np.min(Nt[1:, 0])   # skip the first time since t=0
    tmax = np.max(Nt[1:, 0])   # skip the first time since t=0
    
    log_times = gen_log_times(tmin, tmax, npts)
    index_array = find_argmins(Nt, log_times)
    
    return index_array

```

```python

def plot_conf(ell0, Nt, npts=20,  wmin=-6, wmax=3, rescaled=False, rescaled_pumping=False, L0=1, tau=1, show_sim=False) :
    global eta, gamma, e0
    theta = np.pi/3.
    mu = calc_mu(theta=theta)
    
    T = (24/mu*np.sin(theta)) * (eta*ell0*L0**3)/(gamma*e0**3)
    
    if show_sim :
        for n in range(nsim) :
            index_array = gen_index_array(Nt[ell0][n], npts)
            if rescaled :
                plt.plot(Nt[ell0][n][index_array, 0]/T, Nt[ell0][n][index_array, 1], color = ca_dict[ca][1], marker=chi_dict[chi][0], markersize=chi_dict[chi][1], alpha = 0.1)
            elif rescaled_pumping :
                plt.plot(Nt[(chis, chiv)][ca][n][index_array, 0]/T_ca, Nt[(chis, chiv)][ca][n][index_array, 1], color = ca_dict[ca][1], marker=chi_dict[chi][0], markersize=chi_dict[chi][1], alpha = 0.1)
            else :
                plt.plot(Nt[(chis, chiv)][ca][n][index_array, 0], Nt[(chis, chiv)][ca][n][index_array, 1], color = ca_dict[ca][1], marker=chi_dict[chi][0], markersize=chi_dict[chi][1], alpha=0.1)

    res_avg, res_std = an.batch(data_dict=Nt[ell0], wmin=wmin, wmax=wmax, nwindow=npts)
    if rescaled :
        plt.plot(res_avg[:, 0]/T, res_avg[:, 1], color = ell_param[ell0][0], marker = ell_param[ell0][1], markersize = ell_param[ell0][2], linestyle = ell_param[ell0][3], linewidth = ell_param[ell0][4], label = r'$\ell_0 = ' + str(ell0) + '$')
    else :    
        plt.plot(res_avg[:, 0], res_avg[:, 1], color = ell_param[ell0][0], marker = ell_param[ell0][1], markersize = ell_param[ell0][2], linestyle = ell_param[ell0][3], linewidth = ell_param[ell0][4], label = r'$\ell_0 = ' + str(ell0) + '$')
    

```
