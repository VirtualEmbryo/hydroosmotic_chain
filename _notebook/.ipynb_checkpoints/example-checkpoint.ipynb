This notebook is illustrating results that can be obtained using the script chain.py in the ressources folder.

We first show how one can run the chain.py script with 2 lumens.
Then we show results on a chain of 10 lumens and finally on 100 lumen chain

```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys

module_path = os.path.abspath(os.path.join('../'))

if module_path not in sys.path :
    sys.path.append(module_path)

import _ressources.chain as chain
import _ressources.tools as tools
```

```python
main_dir = '/Users/mathieuleverge/git/hydroosmotic_chain/_data'
```

# 2 Lumens

```python
subdir='2lumens'
conf, ch = chain.load_config(os.path.join(main_dir, subdir, 'config.conf'))
x = np.linspace(0., ch.total_length, 1001)
tools.plot_profile(x, ch, show=True)
print(ch)
```

```python
max_step = int(conf['integration']['max_step'])
alpha = float(conf['integration']['alpha'])
tolerance = float(conf['integration']['tolerance'])
solver = conf['integration']['solver']
nb_frames = int(conf['sim']['nb_frames'])
recording = eval(conf['sim']['recording'])
savefig = eval(conf['sim']['savefig'])

dir_name = os.path.join(main_dir, subdir)
chain.run(ch, max_step=max_step, alpha=alpha, recording=recording, tolerance=tolerance, nb_frames=nb_frames, solver=solver, savefig=savefig, state_simul=True, dir_name=dir_name, pics_dirname=os.path.join(dir_name, 'pics'))
```

```python
ell = tools.load_brfile(os.path.join(main_dir, subdir, 'sim_bridges.dat'))

x_log = 1

L, nions, pos = tools.load_file(os.path.join(main_dir, subdir, 'sim_all.dat'), hydroosmotic=True)
tools.plot_evolution(L, nions, ell, x_logscale=x_log)
```

# 100 Lumens

```python
subdir='100lumens'
conf, ch = chain.load_config(os.path.join(main_dir, subdir, 'config.conf'))
x = np.linspace(0., ch.total_length, 1001)
tools.plot_profile(x, ch, show=True)
#print(ch)

ell = tools.load_brfile(os.path.join(main_dir, subdir, 'sim_bridges.dat'))

x_log = 1

L, nions, pos = tools.load_file(os.path.join(main_dir, subdir, 'sim_all.dat'), hydroosmotic=True)
tools.plot_evolution(L, nions, ell, x_logscale=x_log)
```

```python
Nt = np.loadtxt(os.path.join(main_dir, subdir, 'sim_nlum.dat'))

plt.figure(figsize=(6,6))
plt.xscale('log')
plt.yscale('log')

chiv = 500.
ell0, L0 = 10., 1.
mu, eps = ch.lumens_dict[1].mu, ch.lumens_dict[1].eps
xiv = np.sqrt(L0* float(chiv)**2 * ell0**3)

Th = 2 * ch.tauv * ell0 * L0 / (mu*eps * xiv**2) # Rescaling time

plt.plot(Nt[:, 0]/Th, Nt[:, 1], color = 'red', label='Data')

k, a, t = 100, -2./5, np.logspace(0, 4)
plt.plot(t, k*t**a, linestyle='--', color='k', label = r'$t^{-2/5}$')

plt.xlabel(r'Rescaled time $t / T_h$', fontsize=15)
plt.ylabel('Number of lumens N(t)', fontsize=15)
plt.title('Number of lumens versus time for \n'+r'$\chi_v=500$, $\chi_s=5$, $\tau_{s,v}=1s$', fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

Remark : the hydraulic time $T_h$ is defined as
$$T_h = \frac{2\tau_v \ell_0 L_0}{\mu \epsilon \xi_v^2}$$
where 
$$\xi_v^2 = \ell_0 L_0 \bar{\xi}^2_v$$
by definition of $\chi_v = \bar{\xi}_v / \ell_0$, one gets
$$\xi_v^2= L_0 \chi_v^2 \ell_0^3$$
