```python
import numpy as np
import matplotlib.pyplot as plt

import os
import sys

module_path = os.path.abspath(os.path.join('..'))

if module_path not in sys.path :
    sys.path.append(module_path)

    
import _ressources.chain as chain
import _ressources.tools as tools
import _ressources.lumenclass as lc

main_dir = '/Users/mathieuleverge/git/chain_lumen/'
data_dir = '_data'
```

# Osmotic Equilibrium


```python
show_chain = False
show_area = 1
savefig = 0
x_log = 0
```

## No Screening

```python
path1 = os.path.join(main_dir, data_dir, 'competition_osmotic/osmotic_eq', 'no_screening')

L_ns, nions_ns, pos_ns = tools.load_file(os.path.join(path1, 'sim_all.dat'))
ell_ns = tools.load_brfile(os.path.join(path1, 'sim_bridges.dat'))

conf, ch = chain.load_config(os.path.join(path1, 'config.conf'))

if show_chain :
    print(ch)
    x = np.linspace(0, ch.total_length, 1001)
    tools.plot_profile(x, ch, centers=0, lw=1.5, show=True)
```

```python
savename_ns = os.path.join(main_dir, '_data/competition_osmotic/osmotic_eq/no_screening/graphs.eps')

tools.plot_evolution(L_ns, nions_ns, ell_ns, savename=savename_ns, show_totalarea=show_area, savefig=savefig, x_logscale=x_log, figsize=(8, 8))
```
