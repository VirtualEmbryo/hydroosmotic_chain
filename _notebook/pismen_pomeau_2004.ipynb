# I - Basic Equations
## A - Lubrication equations

$$\mu_s = \frac{Q_s}{h^3} \left[ 1 - \left(\frac{h_m}{h}\right)^3\right]$$

$\mu_s(h)$ : disjoining potential

$Q_s$ : difference between characteristic fluid-fluid and fluid-solid interaction energies

$h_m$ : precursor thickness


```python
def disjoining_potential(hbar) :
    return 1 - 1./hbar**3

hbar = np.linspace(1., 4., 1001)

plt.plot(hbar, disjoining_potential(hbar))
```

```python
X_vec = np.array([1., 5.])    # positions
a_vec = np.array([0.5, 0.1]) # radii

```

```python
def profile(x, a_vec, X_vec, h_m = 1e-5) :
    h = 1.
    for k in range(len(a_vec)) :
        a_k = a_vec[k]
        r_k = np.abs(x - X_vec[k])
        h_k = h_m*(1. + h_m/a_k)
        h *= (a_k / r_k)**(h_k-h_m)
    return np.log(h)
```

```python
x = np.linspace(0., 10., 1001)
h_profile = profile(x, a_vec, X_vec)
```

```python
plt.plot(x, h_profile)
```
